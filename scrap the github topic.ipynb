```python
 import requests  #Use the requests library to download web pages
 from bs4 import BeautifulSoup  # Python library for pulling data out of HTML and XML files
 import pandas as pd  # to create the csv file  
 import os
```

## creating the topic csv file 

```python
topics_url = 'https://github.com/topics'  #url of the web page that we going to do scrapping 
base_url =  'https://github.com' # base url 
```

```python
response = requests.get(topics_url)
```

```python
page_contents = response.text
page_contents[:1000]
```

```python
with open('webpage.html', 'w', encoding='utf-8') as f:
    f.write(page_contents)
```

```python
doc = BeautifulSoup(page_contents , 'html.parser')
```

```python
doc
```

```python
#grabing the topics title 

selection_class = 'f3 lh-condensed mb-0 mt-1 Link--primary'
topics_tittle_tags = doc.find_all('p', {'class': selection_class})
topics_tittle_tags
```

```python
#topic description

dec_class = 'f5 color-fg-muted mb-0 mt-1'
topics_dec_tags = doc.find_all('p', {'class': dec_class})
topics_dec_tags
```

```python
#grabbing the link (we need to go to the topic pages to get the repositary details)

topic_link_tags = doc.find_all('a',{'class': 'no-underline flex-grow-0'})

for tag in topic_link_tags:
       print(tag['href'])
```

```python
# creating the list of full page urL 

topic0_url = []
for tag in topic_link_tags:
    topic0_url.append(base_url + tag['href'])
print(topic0_url)  
```

```python
#creating the list of the topic title 

topic_title  = [] 
for tag in topics_tittle_tags:
   topic_title.append(tag.text)
print(topic_title)
```

```python
# creating a list of the topic description 

topic_discription  = [] 

for tag in topics_dec_tags:
   topic_discription.append(tag.text.strip())
print(topic_discription)
```

```python
#creating topic data frame 

topic_df  = pd.DataFrame({'topic_title':topic_title, 'topic_discription':topic_discription,'topic_url':topic0_url})
topic_df
```

```python
#convert into csv file 

topic_df.to_csv('topics.csv',index=None)
```

## getting information from subtopic and creatins csv file  (for one sub topic)

```python
topic_page_url = topic0_url[0]
topic_page_url
```

```python
response = requests.get(topic_page_url)
response.status_code
```

```python
topic_doc = BeautifulSoup(response.text,'html.parser')
```

```python
# convet star count into a integer to that in hear we create a function 

def parse_star_count(stars_str):
    stars_str = stars_str.strip()
    if stars_str[-1] == 'k':
        return int(float(stars_str[:-1]) * 1000)
    return int (stars_str)
```

```python
#creating the star count list 

star_count_list = []
star_tags = topic_doc.find_all('span',{'class':'Counter js-social-count'})
for tags in star_tags :
    star_count_list.append(parse_star_count(tags.text.strip()))
    
print(star_count_list)
```

```python
#getting the username and repositary name 
    
h3_selection_class = 'f3 color-fg-muted text-normal lh-condensed'   
repo_tags = topic_doc.find_all('h3',{'class': h3_selection_class })
```

```python
def get_repo_details(repo_tags):
    username_list = []
    reponame_list = []
    repo_url_list = []
    for tags in repo_tags:
        a_tags = tags.find_all('a')
        reponame_list.append(a_tags[1].text.strip())
        username_list.append(a_tags[0].text.strip())
        repo_url_list.append('https://github.com' + a_tags[1]['href'])
    
    
    dec1 = {'username':username_list ,'repo name':reponame_list,'repo url':repo_url_list}
    return dec1
```

```python
sub_topic_df = pd.DataFrame(get_repo_details(repo_tags))
sub_topic_df
```

```python
sub_topic_df['star count'] = star_count_list
sub_topic_df
```

```python
# convert into csv file 
sub_topic_df.to_csv('sub topics.csv',index=None)
```

## to get the csv fils for all topics 

```python
def get_sub_topic_doc(topics_url):
    topic_page_url = topics_url
    response = requests.get(topic_page_url)

    if response.status_code != 200:
        raise Exception('Failed to load page {}'.format(topic_page_url))

    topic_doc = BeautifulSoup(response.text, 'html.parser')

    return topic_doc

def get_sub_topic_dataframe (topic_doc) :
    
    star_count_list = []
    star_tags = topic_doc.find_all('span',{'class':'Counter js-social-count'})
    for tags in star_tags :
         star_count_list.append(parse_star_count(tags.text.strip())) 
            
    
    h3_selection_class = 'f3 color-fg-muted text-normal lh-condensed'   
    repo_tags = topic_doc.find_all('h3',{'class': h3_selection_class }) 
    
    
    username_list = []
    reponame_list = []
    repo_url_list = []
    for tags in repo_tags:
        a_tags = tags.find_all('a')
        reponame_list.append(a_tags[1].text.strip())
        username_list.append(a_tags[0].text.strip())
        repo_url_list.append('https://github.com' + a_tags[1]['href'])
    
    
    dec1 = {'username':username_list ,'repo name':reponame_list,'repo url':repo_url_list, 'star count' :star_count_list }
    
    
    return dec1   
    
    
            
    
    
    
    
    
    
```

```python
def scraping_topics ( topic_title,topic0_url) :
    fname = topic_title + '.csv'
    if os.path.exists(fname):
        print ("the file {} alresdy exsits. Skipping ....".format(fname) )
        return
    sub_topic_df = pd.DataFrame(get_sub_topic_dataframe (get_sub_topic_doc(topic0_url)))
    sub_topic_df.to_csv(fname,index=None)
    
```

```python
for index, row in topic_df.iterrows():
    scraping_topics(row['topic_title'],row['topic_url'])
    
    
```
